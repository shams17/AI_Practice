```python
#histogram: continous numerical univariate variables

%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets

iris_data = datasets.load_iris()

print(type(iris))
```

```python
#needs to be converted to data frame
import pandas as pd
iris = pd.DataFrame(data=np.c_[iris_data['data'], iris_data['target']], columns=iris_data['feature_names']+['species'])
iris.head()
```

```python
#replace 0 and 1 entries with setosa and versicolor
'''
🧠 What it means:
iris.species initially contains numbers like 0.0, 1.0, 2.0.

You want to replace those numbers with the actual flower names:

0.0 → "setosa"

1.0 → "versicolor"

2.0 → "virginica"

🧩 How np.where() works here:
First np.where(iris.species == 0.0, "setosa", ...)

If a value is 0.0, **replace with "setosa".

Otherwise, move to the next np.where() inside.

Then np.where(iris.species == 1.0, "versicolor", "virginica")

If a value is 1.0, **replace with "versicolor".

Otherwise (meaning it must be 2.0), **replace with "virginica".

🛠 So in simple words:
Before (species)	After (species)
0.0	setosa
1.0	versicolor
2.0	virginica
'''
iris.species = np.where(iris.species == 0.0, 'setosa', np.where(iris.species == 1.0,'versicolor', 'virginica'))
iris.head(n=5)
```

```python
#Boxplots: visualize the distribution of continous numerical variables
#helps us to visualize the min, max, median and how the data is distributed
'''
📚 What is the Median?
The median is the middle value in a set of numbers when they are sorted in order.

It is NOT affected by extreme values (outliers) as much as the mean (average) is.

It gives a better sense of the "central tendency" when data are skewed.

🧠 Step-by-Step:
Arrange the numbers in ascending order (smallest → largest).

Pick the middle value:

If odd number of items → median is the middle item.

If even number of items → median is the average of the two middle items.

🎯 Example 1: Odd Number of Items
Suppose your data is:

text
Copy code
5, 2, 9, 1, 7
First, sort them:

text
Copy code
1, 2, 5, 7, 9
Number of items = 5 (odd)

Middle value = 3rd item = 5

✅ So the median is 5.

🎯 Example 2: Even Number of Items
Suppose your data is:

text
Copy code
8, 3, 7, 5
First, sort them:

text
Copy code
3, 5, 7, 8
Number of items = 4 (even)

Middle two values = 5 and 7

Now, median = (5 + 7) / 2 = 6

✅ So the median is 6.

📢 Why is Median Important?
✅ Median resists outliers:

Imagine incomes:

text
Copy code
30K, 32K, 35K, 38K, 1 Million
Mean (average) = extremely high (skewed by 1 million).

Median = middle value = 35K → gives a realistic center.

Thus, in skewed data, median gives a better sense of reality.


'''
iris.head(n=5)
```

```python
#calculating the standard deviation to calculate the variance
'''
📚 What is Standard Deviation?
Standard deviation is a measure of how spread out the numbers are in a dataset.

It tells you:

Are the values close to the average (mean)?

Or are they widely scattered?

🎯 Intuition
Imagine two classes took the same test:

Class A scores: 80, 81, 79, 82, 80

Class B scores: 60, 90, 50, 100, 70

➡️ Both classes have the same average (~80)
But Class A is tightly grouped around the mean → low standard deviation
Class B is spread out → high standard deviation

🧠 Definition (Math)
Standard deviation is the square root of variance:

text
Copy code
Standard Deviation (σ) = √(Σ(xᵢ − μ)² / N)
Where:

xᵢ = each data point

μ = mean of data

N = number of points

✅ Simple Example
Data:

text
Copy code
[10, 12, 13, 14, 15]
Mean = 12.8

Differences from mean = [-2.8, -0.8, 0.2, 1.2, 2.2]

Squared differences = [7.84, 0.64, 0.04, 1.44, 4.84]

Variance = average of those = 2.96

Standard deviation = √2.96 ≈ 1.72

So, most values are within ~1.7 units of the mean.

🛠️ In Python:
python
Copy code
import numpy as np

data = [10, 12, 13, 14, 15]
print(np.std(data))  # Output: ~1.72
📊 Why is it important?
Low standard deviation → data is consistent

High standard deviation → data is spread out or volatile

📌 Use Cases
Domain	Use of Standard Deviation
Finance	To measure stock volatility (risk)
Manufacturing	To measure consistency in product size
Education	To compare performance variation in exams
Machine Learning	Feature scaling, outlier detection
🚀 Summary
Term	Meaning
Standard Deviation	Spread or dispersion from the mean
Low std	Data clustered near mean (tight)
High std	Data is more spread out (loose)
'''
sd = iris.groupby(['species']).std()
sd.head()
```

```python
import csv
import random
import os

# Define the output directory and filename
output_dir = r"C:/Users/hebbarsh/Desktop/IntelliLearn/DataVisualizarioAndWrangling/Dc/DataAndCode/section8"
output_file = "unique_countries_expenditure.csv"
output_path = os.path.join(output_dir, output_file)

# Create the directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# List of 50 unique countries
countries = [
    "USA", "Canada", "UK", "Germany", "France", "Japan", "Australia", "India", "Brazil", "South Africa",
    "Mexico", "Russia", "Italy", "Spain", "Netherlands", "Sweden", "Norway", "Denmark", "Finland", "Poland",
    "Argentina", "Chile", "Colombia", "Peru", "Venezuela", "Egypt", "Nigeria", "Kenya", "Morocco", "Turkey",
    "Saudi Arabia", "UAE", "Israel", "South Korea", "China", "Thailand", "Vietnam", "Malaysia", "Indonesia",
    "Philippines", "New Zealand", "Ireland", "Portugal", "Belgium", "Switzerland", "Austria", "Greece", "Czech Republic", "Hungary"
]

institution_types = ["University", "Government", "Private"]
expenditure_types = ["Research", "Infrastructure", "Operations"]

header = [
    "country",
    "institution_type",
    "direct_expenditure_type",
    "expenditure_1995",
    "expenditure_2000",
    "expenditure_2005",
    "expenditure_2010",
    "expenditure_2011",
]

def generate_expenditures(base):
    expenditures = []
    for year in [1995, 2000, 2005, 2010, 2011]:
        if year == 1995:
            val = base
        elif year == 2011:
            val = int(expenditures[-1] * 1.02)
        else:
            val = int(expenditures[-1] * 1.05)
        expenditures.append(val)
    return expenditures

data = []
for country in countries:
    institution = random.choice(institution_types)
    expenditure_type = random.choice(expenditure_types)
    base_expenditure = random.randint(10000, 70000)
    expenditures = generate_expenditures(base_expenditure)
    row = [country, institution, expenditure_type] + expenditures
    data.append(row)

# Write CSV file
with open(output_path, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data)

print(f"CSV file created successfully at:\n{output_path}")
```

```python
#get the count of number of similar category prizes were won that year
'''
🧠 Step-by-Step Explanation:
1. data.groupby(['Year', 'Category'])
This groups the DataFrame data by:

Year: the year the award was given

Category: like Peace, Physics, Chemistry, etc.

Each combination of Year and Category forms a group.

2. ['Laureate ID'].count()
For each (Year, Category) group, this counts how many laureates there were.

It does this by counting non-null entries in the 'Laureate ID' column.

So, it's basically: “How many unique people received awards in each year and category?”

3. .reset_index()
After groupby().count(), you get a special object called a GroupBy DataFrame with a multi-index (Year, Category).

Calling .reset_index():

Flattens it back into a normal DataFrame

Makes Year, Category, and Laureate ID regular columns

🧾 The Resulting DataFrame year_cat looks like:
Year	Category	Laureate ID
1901	Physics	1
1901	Chemistry	1
1902	Physics	2
...	...	...
✅ So now you can easily plot, analyze, or filter by year and category.

📊 What This Is Useful For:
See trends in how many people won each year by category

Build a bar chart or line chart over time

Identify anomalies (e.g., no Peace Prize awarded some year)
'''
year_cat = data.groupby(['Year', 'Category'])['Laureate ID'].count().reset_index()
year_cat
```

```python
'''
🎯 Objective:
You're trying to:

Create small multiples (i.e., a separate plot for each Category)

Show how the number of Laureates per category has changed over the years

Use both line plot and shaded area under the curve for visual clarity

🧠 Step-by-Step Breakdown
🔹 sns.FacetGrid(year_cat, col='Category', hue='Category', col_wrap=4)
FacetGrid creates a grid of subplots (one for each category).

col='Category': Each subplot represents a unique Nobel category (e.g., Peace, Physics).

hue='Category': Assigns a different color to each category.

col_wrap=4: Arrange the plots in rows with maximum 4 columns per row.

📌 At this stage, the layout is ready but no plots are drawn yet.

🔹 g = g.map(plt.plot, 'Year', 'Laureate ID')
This maps a line plot onto each subplot:

X-axis: Year

Y-axis: Laureate ID (actually the count of laureates per year)

It draws one line per subplot for each category.

🔹 g = g.map(plt.fill_between, 'Year', 'Laureate ID', alpha=0.2)
This fills the area under the line (shaded region):

alpha=0.2 makes the fill slightly transparent.

This helps emphasize the shape of the trend.

🔹 .set_titles("{col_name} Category")
Sets the title of each subplot to the category name (e.g., Physics Category)

{col_name} is a placeholder replaced with the actual category.

🔹 g = g.set_titles("{col_name}")
Optionally simplifies the title to just the category name (e.g., Physics).

You might use either this or the one above — not both necessarily.

📊 Visualization Summary:
This code gives you a grid of line charts, each showing:

How the number of laureates has changed over the years

For each Nobel category

With shaded area under the curve for visual clarity

✅ Example Output (Mentally Picture):
matlab
Copy code
+------------+------------+------------+------------+
|  Physics   | Chemistry  | Medicine   | Peace      |
|  (plot)    |  (plot)    |  (plot)    |  (plot)    |
+------------+------------+------------+------------+
|  Economics | Literature | ...        | ...        |
+------------+------------+------------+------------+

'''
g = sns.FacetGrid(year_cat, col='Category', hue='Category', col_wrap=4, )
g = g.map(plt.plot, 'Year', 'Laureate ID')
g = g.map(plt.fill_between, 'Year', 'Laureate ID', alpha=0.2).set_titles("{col_name} Category")
g = g.set_titles("{col_name}")
plt.show()
```
