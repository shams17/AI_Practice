```python
#univariate feature selection
'''
🌟 What is Univariate Feature Selection?
"Uni" means one.

In univariate feature selection, we evaluate each feature individually (one by one) against the target variable to decide how useful it is.

You do NOT look at how features interact with each other — just feature vs target separately.

🔵 Main goal:
Pick the best features based on their own individual relationship with the target (output variable).

🌟 How does it work?
Choose a scoring function that measures how strong the relationship is.

Example scoring functions:

For classification (target has classes like 0/1, cat/dog):

chi2 (Chi-squared test)

f_classif (ANOVA F-test)

mutual_info_classif (mutual information)

For regression (target is continuous like prices, scores):

f_regression

mutual_info_regression

Then rank features based on these scores.

Finally select top features.

✅ You are reducing features — keeping only the most useful ones!

🌟 Example (with code!)
Imagine this simple data:

python
Copy code
import pandas as pd

# Sample data
data = pd.DataFrame({
    'height': [150, 160, 165, 170, 180],
    'weight': [50, 60, 65, 70, 80],
    'age': [25, 30, 35, 40, 45],
    'blood_type': [1, 2, 1, 2, 1],  # blood type coded numerically
    'target': [0, 1, 0, 1, 0]  # classification target
})
We want to select best features to predict target.

🌟 Doing Univariate Feature Selection
python
Copy code
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = data[['height', 'weight', 'age', 'blood_type']]  # Features
y = data['target']  # Target

# Step 1: Apply univariate feature selection
selector = SelectKBest(score_func=chi2, k=2)  # Select top 2 features
X_new = selector.fit_transform(X, y)

# Step 2: See which features were selected
mask = selector.get_support()  # [True, False, True, False]
selected_features = X.columns[mask]
print(selected_features)
✅ Output:

pgsql
Copy code
Index(['height', 'age'], dtype='object')
So, based on the chi2 test, height and age are most useful to predict the target.

🌟 Visual flow
nginx
Copy code
height  ───> compute score
weight  ───> compute score
age     ───> compute score
blood   ───> compute score

Rank features → Pick top 2 → Keep 'height' and 'age'
🌟 Why use Univariate Feature Selection?
Speeds up your model (fewer columns)

Removes noise (useless features)

Reduces overfitting risk

Easy to understand and implement

Works well when individual features are strongly predictive

🌟 Important:
Univariate ignores feature interactions.
(maybe 2 bad features together could be powerful — but univariate won't catch that).

So sometimes after univariate selection, we also apply other feature engineering techniques.

🎯 When to use:
Scenario	Is Univariate Good?
Many features (>100)	✅ Yes, filter quickly
Small dataset	✅ Yes
You want fast simple filtering	✅ Yes
You need feature combinations to matter	❌ No, use other methods
🌟 Real Life Example
Imagine predicting whether a patient has diabetes.

Features	
Blood sugar	🩸 Very relevant
Eye color	👁️ Not relevant
BMI	✅ Relevant
Phone brand	📱 Not relevant
You would want univariate feature selection to keep Blood sugar and BMI, and remove Eye color and Phone brand.

🧠 Small checklist while doing Univariate Feature Selection:
Split data into X (features) and y (target).

Choose a scoring function depending on task (classification or regression).

Use SelectKBest (or similar) to pick top features.

Train your model with only selected features.

🔥 Final important point:
Choosing k (how many features to keep) is important!

You can manually set k=5 (keep top 5).

OR you can cross-validate and pick the number giving the best model performance.
'''
import numpy as np #linear algebra

import seaborn as sns #data visulaization library
import matplotlib.pyplot as plt

from sklearn.feature_selection import RFE

breast = pd.read_csv('C:/Users/hebbarsh/Desktop/IntelliLearn/DataVisualizarioAndWrangling/Dc/DataAndCode/section5/breast.csv')
breast.head(n=7)

```

```python
#this is a little annoying to manually see which features has the best scores and mapping it here
#better thing is use another feature selection technique which recursive feature selection:
'''
🌟 What is RFE (Recursive Feature Elimination)?
RFE is a feature selection method.

It automatically finds the most important features for a model.

It recursively removes least important features, one by one (or few at a time), until the desired number of features is left.

✅ It’s a bit like peeling an onion layer by layer — you keep removing the least useful parts.

🌟 Why use RFE?
Some features are not helping (maybe they are noisy or redundant).

Removing weak features makes models faster, simpler, and sometimes even more accurate.

RFE selects features based on how well they help the model, not just correlation.

🌟 How does RFE work (internally)?
Imagine you have 6 features: [A, B, C, D, E, F]

RFE steps:

Train the model (say Logistic Regression) on all features.

Rank features by importance.

Remove the least important feature (say, F).

Train the model again on [A, B, C, D, E].

Again, rank and remove the next least important feature.

Repeat until the desired number of features is left (say 3).

🌟 RFE Code Example (Simple)
Let's work with a small dataset 🛠️.

python
Copy code
# Import libraries
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE

# Load sample data
data = load_breast_cancer()
X = data.data
y = data.target

# Initialize model
model = LogisticRegression(max_iter=5000)

# Apply RFE
rfe = RFE(estimator=model, n_features_to_select=5)  # Select top 5 features
X_rfe = rfe.fit_transform(X, y)

# Which features got selected?
selected = rfe.support_
selected_features = data.feature_names[selected]
print(selected_features)
✅ Output:

css
Copy code
['mean radius' 'mean texture' 'mean perimeter' 'mean area' 'mean concavity']
👉 So RFE decided these 5 features are best for predicting cancer outcome!

🌟 Visual explanation
sql
Copy code
Initial features:  ['A', 'B', 'C', 'D', 'E', 'F']
Train model → Rank feature importance
Remove 'F'
Train model → Rank again
Remove 'E'
...
Final features: ['A', 'B', 'C']
🌟 More Control in RFE
You can control:

estimator — what model is used (Logistic Regression, Random Forest, SVM, etc).

n_features_to_select — how many features you want to keep.

step — how many features to remove at a time.

Example of removing 2 features at a time:

python
Copy code
rfe = RFE(estimator=model, n_features_to_select=5, step=2)
🌟 When to use RFE?
When	Why
You want the best k features	✅ RFE is smart
Features are not clearly correlated	✅ RFE understands feature interactions
Small to medium datasets	✅ Because RFE retrains many times
Very large datasets (>100K samples)	❌ Slow. Use faster methods
🌟 Important: Which Model to use inside RFE?
Model inside RFE	Best for
Logistic Regression	Classification tasks
Linear Regression	Regression tasks
Decision Tree / Random Forest	When relationships are complex (non-linear)
SVM (Support Vector Machine)	High-dimensional data
You can choose the estimator based on your task!

🌟 RFE vs Univariate Selection
RFE	Univariate
Looks at feature combinations	✅ Yes	❌ No
More powerful	✅	🚫
Slower (trains multiple models)	🐢	🐇 Fast
Needs a model inside	✅	❌
🌟 A small realistic Example:
Suppose you're predicting House Prices using:

Feature	
Number of bedrooms	🛏️
Area in sq ft	📏
Distance to city center	🚗
Wall color	🎨
Number of kitchens	🍽️
RFE may discover that:

Wall color is not important.

Number of kitchens and Area are super important.

✅ So it would keep Area and Bedrooms, remove Wall color.

✨ Bonus Tip: How to automatically find best number of features?
You can combine RFE + Cross Validation (RFECV)!

Example:

python
Copy code
from sklearn.feature_selection import RFECV

rfecv = RFECV(estimator=model, step=1, cv=5, scoring='accuracy')
rfecv.fit(X, y)

print("Optimal number of features: ", rfecv.n_features_)
🚀 It will auto-select the number of features giving highest validation accuracy!

🧠 In Short:
RFE = Recursively remove worst features until the best ones are left.

Choose an estimator carefully.

Use it to improve models and simplify datasets.

Best for small-medium datasets!

'''
breast.head()
```

```python
#principal component analysis:
'''
🌟 What is PCA (Principal Component Analysis)?
PCA is a Dimensionality Reduction technique.

It transforms a large set of correlated features into a smaller set of uncorrelated components.

These new features (called Principal Components) capture most of the information (variance) in the original data.

✅ Goal: Keep important information, but reduce the number of features.

🌟 Why use PCA?
Problem	How PCA helps
Too many features (curse of dimensionality)	📉 Reduce them
Features are correlated (multicollinearity)	🔄 Create independent components
Visualization (high-dimensional to 2D or 3D)	🎨 Make beautiful plots
Speed up machine learning models	⚡ Faster training
🌟 How PCA works — Step-by-Step Intuition
Suppose you have 2 features:

Height (in cm)

Weight (in kg)

(They are somewhat correlated: taller people often weigh more.)

🔵 PCA steps:

Standardize the data
→ (mean = 0, standard deviation = 1 for each feature)

Find the direction where the data varies the most
→ (First Principal Component)

Find the second direction (orthogonal to the first) with the second highest variance
→ (Second Principal Component)

Project the original data onto these directions
→ (New transformed data)

✅ Now, instead of "height" and "weight", we have "PC1" and "PC2".

🌟 PCA Visualization (imagine):
mathematica
Copy code
Before PCA:
- Axis X = Height
- Axis Y = Weight
(Features correlated)

After PCA:
- Axis PC1 = "Size" (captures both height and weight together)
- Axis PC2 = "Residual" (small variation left)
🌟 Real Example using Python
Let's use Breast Cancer Dataset again.

python
Copy code
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load data
data = load_breast_cancer()
X = data.data
y = data.target

# Step 1: Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components
X_pca = pca.fit_transform(X_scaled)

# Step 3: Visualize
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='rainbow', alpha=0.7)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA - Breast Cancer Dataset')
plt.show()
✅ Output:
You'll see two colorful clouds (Malignant vs Benign tumors) separated nicely by just 2 features (PC1, PC2) instead of original 30 features!

🌟 What is a Principal Component exactly?
A Principal Component is a linear combination of your original features.

Example:

ini
Copy code
PC1 = 0.5 * (height) + 0.7 * (weight)
The numbers (like 0.5, 0.7) are loadings (how much each original feature contributes).

✅ PCA creates new axes in the feature space.

🌟 How many Principal Components should you keep?
Use Explained Variance Ratio:

python
Copy code
print(pca.explained_variance_ratio_)
✅ Example output:

csharp
Copy code
[0.45, 0.18, 0.08, 0.05, ...]
Meaning:

PC1 captures 45% of total variance

PC2 captures 18%

PC3 captures 8%

and so on...

You can plot a Scree Plot (variance vs number of components) and choose components that explain ~95% of total variance.

🌟 Important Points to remember
PCA rotates the feature space	🧩
PCA components are uncorrelated	🔗 ❌
PCA loses interpretability (you can't easily say what PC1 means)	😵
PCA is unsupervised (does not look at target variable)	🚫 🎯
Always standardize features before applying PCA	⚡
🌟 Summary in One Line:
PCA transforms high-dimensional, correlated data into fewer, uncorrelated dimensions that still capture most of the original information.

🌟 Bonus: How PCA works mathematically (optional)
PCA is based on Eigenvalues and Eigenvectors of the Covariance Matrix.

Eigenvectors → Directions (principal components)

Eigenvalues → Strength of the variance in those directions.

🚀 In short:
PCA = Reduce features smartly.

Helps in speeding models, reducing overfitting, and making cool 2D/3D visualizations.

Easy to use in Python with scikit-learn.
'''
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

iris = sns.load_dataset("iris")
iris.head(n=4)
```

```python
#PCA works only on numerical values hence creating a subset with only numerical values
num = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
#convert the numerical values into numpy arrays
x= num.values
x
```

```python
#scaling the values
X = scale(x) #makes sure the array only has values between 0 and 1 since PCA gets skewed by extreme data
pca = PCA(n_components=4)
pca.fit(X)
```

```python
#the amount of variance each PC explains
var = pca.explained_variance_ratio_
var
```

```python
plt.plot(var)
```

```python
#cumilative variance explaination
var1 = np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
plt.plot(var1)
```
