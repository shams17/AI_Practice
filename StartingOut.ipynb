```python
#pandas library converts data from variety of external sources into panda series and panda data frames upon which many operations can be performed
#data series is nothing but labelled 1D array
#dataFrames are 2D arrays
#creating data series from dictinoaries(key: value pairs)
import pandas as pd
dict = {'cat': 'Max Verstappen', 'dog': 'Lando Norris', 'food': 'Oscar piastri', '4': 33}
oneD = pd.Series(dict)
#subscript operator
print(oneD.loc[['cat', 'dog']])
print(oneD['cat'])
#index as a number for an array starting from index 0
print(oneD.iloc[[1, 2]])
print('cat' in oneD)
print('Cat' in oneD)
```

```python
#data Frames = 2D arrays
D = {'A': pd.Series(['100', '200', 300], index=['apple', 'bannana', 'pear']),
     'B': pd.Series(['101', '201', '301', '401'], index=['apple', 'bannana', 'pear', 'melon'])}
df = pd.DataFrame(D)
print(df)
#just print out the index values
print(df.index)
#print out the columns
print(df.columns)
#create a new data frame from existing data frame
print(pd.DataFrame(df, index = ['apple', 'bannana'], columns=['A']))
```
